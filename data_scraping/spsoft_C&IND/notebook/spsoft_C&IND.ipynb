```python
with open(r'C:\Users\GPU\Desktop\data_scraping\spsoft_C&IND\spsoft_C&IND.html','r',encoding='utf-8') as f:
    html = f.read() 
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python

soup = BeautifulSoup(html, "lxml")
```

```python
soup
```

```python
social = soup.find_all('div',{'class':'socialLinks justify-content-end position-fixed sideSocialLinks'})
```

```python
links_df =[]
for i in social:
    links = i.find_all('a')
    for link in links:
        links_df.append(link['href'])
```

```python
links_df
```

```python
link_handles = ['Facebook','X','Linkedin','Youtube','Clutch']
```

```python
social_media = pd.DataFrame([link_handles,links_df]).T
```

```python
social_media.columns=['Handles','Link']
```
