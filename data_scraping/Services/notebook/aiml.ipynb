```python
from bs4 import BeautifulSoup
import pandas as pd

with open(r'C:\Users\GPU\Desktop\data_scraping\Services\aiml_services.html','r',encoding='utf-8') as f:
    html = f.read()
soup = BeautifulSoup(html)
```

### AI projects Portfolio

```python
llm = []
llm1 = []
LLM_data = soup.find_all('div',{'class':'col-md-9 col-lg-9 col-12 learnContent pt-3'})
for i in LLM_data:
    llm.append(i.text)
for i in LLM_data:
    for j in i.find_all('li'):
        llm1.append(j.text)
llm = llm + llm1
llm = [' '.join(llm)]


NLP = []
NLP_points = []
nlp = soup.find_all('p',{'class':'learnContent'})
nlp_points = soup.find_all('ul',{'class':'learnContent mb-0'})
for i in nlp:
    NLP.append(i.text)
NLP1 = [NLP[0]]
for j in nlp_points:
    for k in j:
        NLP_points.append(k.text)
NLP = NLP1 + NLP_points
NLP = [' '.join(NLP)]
```

```python
title = ['Large Language Models','Natural Language Processing']
t_des = llm+NLP
```

```python
project = soup.find_all('div',{'class':'card-title'})
description = soup.find_all('p',{'class':'text-justify text-secondary'})
des = soup.find_all('p',{'class':'card-text text-white mb-0 p-2'})
des1_nlp = soup.find_all('ul',{'class':'learnContent mb-0'})
des1_llm = soup.find_all('div',{'class':'col-md-9 col-lg-9 col-12 learnContent pt-3'})
Project = []
Description = []
Des1 = []
Des1_llm = []
Des1_nlp = []
for i in des1_llm:
    for j in i.find_all('li'):
        Des1_llm.append(j.text)
Des1_llm = [','.join(Des1_llm)]
for i in des1_nlp:
    for j in i.find_all('li'):
        Des1_nlp.append(j.text)
Des1_nlp = [','.join(Des1_nlp)]
Des1_nlp
for i in project:
    Project.append(i.text.strip())
for j in description:
    Description.append(j.text.strip())
for k in des:
    Des1.append(k.text.strip())
Project = Project+title
Description = Description+t_des
Des1 = Des1+Des1_llm+Des1_nlp
```

```python
ai_projects = {
    'Project':Project,
    'Project_Description':Description,
    'Product_solution':Des1
}
AIML_project = pd.DataFrame(ai_projects)
AIML_project
```
