```python
from bs4 import BeautifulSoup
import pandas as pd

with open(r'C:\Users\GPU\Desktop\data_scraping\Services\da_services.html','r',encoding='utf-8') as f:
    html = f.read()
soup = BeautifulSoup(html)
```

### Data Analytics Services

```python
da_hwch = soup.find_all('div',{'class':'d-flex justify-content-center'})
DA_hwch = []
for i in da_hwch:
    DA_hwch.append(i.text)
DA_hwch = {
    'DA(Services)': DA_hwch
}
DA_hwch = pd.DataFrame(DA_hwch)
DA_hwch
```

### Data Analytics Poject Portfolio

```python
da_projects = soup.find_all('div',{'class':'DataAnalytics container-fluid py-5'})
da_des = soup.find_all('p',{'class':'text-justify text-secondary'})
da_des1 = soup.find_all('p',{'class':'card-text text-white mb-0 p-2'})
projects = []
des = []
des1 = []
for i in da_projects:
    for j in i.find_all('h6'):
        projects.append(j.text)
for i in da_des:
    des.append(i.text)
for i in da_des1:
    des1.append(i.text)
DA_projects = {
    'Project':projects,
    'Project_Description':des,
    'Product_solution':des1
}
DA_projects = pd.DataFrame(DA_projects)
DA_projects
```

### Data Analytics Offering

```python
da_off = soup.find_all('p',{'class':'ourOffer'})
da_off_des = soup.find_all('p',{'class':'process'})
DA_off = []
DA_off_des = []
for i in da_off:
    DA_off.append(i.text.strip())
for j in da_off_des:
    DA_off_des.append(j.text.strip())
DA_off_des.pop(0)
DA_off = {
    'Services':DA_off,
    'Description':DA_off_des
}
DA_off = pd.DataFrame(DA_off)
DA_off
```
