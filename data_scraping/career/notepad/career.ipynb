```python
with open(r'C:\Users\ADMIN\Desktop\scrapping\career.html','r',encoding='utf-8') as f:
    html = f.read()
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html)
```

```python
soup
```

```python
career = soup.find_all('h2',{'class':'accordion-header'})
```

```python
job = []
for i in career:
    button = i.find('button')
    if button:
        job.append(button.text.strip())
```

```python
job_des = []
skill = []
edu = []
des = soup.find_all('div',{'class':'accordion-body'})
for i in des:
    job_des.append(i.text.strip().split('Job Description:')[1].split('Skills')[0])
    skill.append(i.text.strip().split('Skills Set:')[1].split('Edu')[0])
    edu.append(i.text.strip().split('Qualification:')[1].split('Send')[0])
```

```python
career = {
    'Vacancy': job,
    'Job_Description':job_des,
    'Skill_Set':skill,
    'Educational_Qualification':edu
}
```

```python
import pandas as pd
career = pd.DataFrame(career)
```
